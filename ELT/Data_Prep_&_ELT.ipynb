```python
#Used Libraries
import pandas as pd
```

```python
#Loading raw_data
raw_data=pd.read_csv("Amazon Sale Report.csv")
raw_data.head(10)
```

```python
#checking nan values
raw_data.isnull().sum()
```

```python
raw_data.shape
```

```python
#Removing dolumns we don't need in our analysis
new_data=raw_data.drop(columns=["SKU","ASIN","ship-postal-code", "promotion-ids", "fulfilled-by", "Unnamed: 22"])
```

```python
new_data.isnull().sum()
```

```python
new_data["ship-country"].unique()
```

```python
#The dataset shows the Amazon sales repost in India. ship-country contains India (IN) only so we can drop it
new_data=new_data.drop(columns=["ship-country"])
```

```python
new_data["currency"].unique()
```

```python
#We can also drop the currecny because it's about the Indian currency only
new_data=new_data.drop(columns=["currency"])
```

```python
#dealing with missing values
new_data=new_data.dropna(subset=['Courier Status'])
```

```python
new_data.isnull().sum()
```

```python
new_data=new_data.dropna(subset=["ship-city","ship-state"])
```

```python
new_data.isnull().sum()
```

```python
#filling Amount missing values with the mean value to not affect the overall data measurements
Mean=new_data["Amount"].mean()
Mean
```

```python
new_data["Amount"]=new_data["Amount"].fillna(Mean)
```

```python
new_data["Amount"].mean()
```

```python
new_data.isnull().sum()
```

```python
new_data.shape
```

```python
#We have filled the Amount missing values with keeping the same mean value
```

```python
new_data.head()
```

```python
new_data.reset_index(drop=True,inplace=True)
```

```python
new_data.head()
```

```python
new_data.drop(columns=["index"],inplace=True)
```

```python
new_data.head()
```

```python
#Saving our dataset
data=new_data.copy()
data.to_csv("Amazon_Data.csv",index=False)
```
